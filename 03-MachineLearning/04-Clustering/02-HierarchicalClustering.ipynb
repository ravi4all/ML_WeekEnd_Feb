```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift
import matplotlib.pyplot as plt
```

```python
centers = [[1,1],[5,5],[3,10]]
X, _ = make_blobs(centers=centers, n_samples=1500, cluster_std=1)

plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
ms = MeanShift()
ms.fit(X)
```

```python
centeroid = ms.cluster_centers_
labels = ms.labels_
```

```python
colors = ["r.","g.","c."]

for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colors[labels[i]])

plt.scatter(centeroid[:,0],centeroid[:,1],
            marker="x",color='k', s=150, linewidths = 5, zorder=10)
plt.show()
```
