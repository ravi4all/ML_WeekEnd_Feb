```python
# Import modules
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([150,200,250,300,400,600,329,440,510])
y = np.array([6450,7450,8450,9450,11450,15450,15150,12000,14450])
```

```python
slope, intercept, r_value, p_value, std_error = stats.linregress(x,y)
```

```python
plt.plot(x,y,'ro', color='black')
plt.xlabel('Size of house')
plt.ylabel('Price')
plt.axis([0,650,0,19000])
plt.plot()
```

```python
plt.plot(x,y,'ro', color='black')
plt.xlabel('Size of house')
plt.ylabel('Price')
plt.axis([0,650,0,19000])
plt.plot(x,x*slope+intercept, 'b')
plt.plot()
```

```python
# Prediction
newX = 650
newY = newX*slope+intercept
print(newY)
```
